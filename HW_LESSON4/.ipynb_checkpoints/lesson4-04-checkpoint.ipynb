```python
from matplotlib import pyplot as plt
import numpy as np
%matplotlib inline
%config InlineBackend.figure_format = 'svg'
import pandas as pd 

df=pd.read_csv('./creditcard.csv')
```

```python
#df10=df.head(10)
#print(df10.iloc[:3,:2])
ser_counts=df['Class'].value_counts()

plt.style.use('fivethirtyeight')

ser_counts.plot(kind='bar')
plt.show()
ser_counts.plot(kind='bar', logy=True)
plt.show()
```

```python
plt.hist(df.loc[df['Class']==1,'V1'], color='red', bins=20,alpha=1, density=True)
plt.hist(df.loc[df['Class']==0,'V1'], color='green', alpha=0.5, bins=20, density=True)
plt.show()
```
